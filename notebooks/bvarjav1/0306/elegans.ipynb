```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import graspy
from tqdm import tqdm
import networkx as nx
%matplotlib inline
```

```python
male = np.array(pd.read_csv('male_chem_A_full_undirected.csv', header=None).values)
herm = np.array(pd.read_csv('herm_chem_A_full_undirected.csv', header=None).values)
```

```python
print('male:',male.shape)
print('herm:',herm.shape)
```

```python
graspy.plot.gridplot([male], ['male'])
```

```python
graspy.plot.gridplot([herm], ['herm'])
```

```python
nonpar = graspy.inference.NonparametricTest()
p = nonpar.fit(male[:50,:50],herm[:50,:50])
plt.hist(nonpar.null_distribution_, 50)
plt.axvline(nonpar.sample_T_statistic_)
plt.title(p)
plt.show()
```

```python
listxy = []
for i in tqdm(range(10)):
    xr = range(0,500,50)
    yr = range(0,400,50)
    xy = np.zeros((len(xr)+1, len(yr)+1))
    i = 0
    j = 0
    for x in xr:
        for y in yr:
            m = male[x:x+50,x:x+50]
            h = herm[y:y+50,y:y+50]
            p = nonpar.fit(m,h)
            xy[i,j] = p
            j += 1
        i += 1

        j = 0
    listxy.append(xy)
```

```python
import seaborn as sns
sns.heatmap(xy)
```

```python
axy = np.array(listxy)
axym = np.mean(axy,axis=0)
axym.shape
```

```python
sns.heatmap(axym)
```
